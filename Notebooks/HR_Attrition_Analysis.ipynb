```python
#creating groups/ categories of features for analysis 

demographic_features = ['Age', 'Gender', 'MaritalStatus', 'Education', 'EducationField']

job_features = ['Department', 'JobRole', 'JobLevel', 'YearsAtCompany', 
                'YearsInCurrentRole', 'YearsSinceLastPromotion', 
                'YearsWithCurrManager', 'TotalWorkingYears', 'NumCompaniesWorked']

satisfaction_features = ['JobSatisfaction', 'EnvironmentSatisfaction', 
                        'WorkLifeBalance', 'RelationshipSatisfaction', 'JobInvolvement']

compensation_features = ['MonthlyIncome', 'MonthlyRate', 'DailyRate', 'HourlyRate',
                        'PercentSalaryHike', 'StockOptionLevel']

work_condition_features = ['OverTime', 'BusinessTravel', 'DistanceFromHome']

performance_features = ['PerformanceRating', 'TrainingTimesLastYear']
```

```python
#creating new groups 

df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 30, 40, 50, 100], 
                        labels=['20-30', '30-40', '40-50', '50+'])

# Create income brackets
df['IncomeGroup'] = pd.cut(df['MonthlyIncome'], bins=3, 
                           labels=['Low', 'Medium', 'High'])

# Create tenure groups
df['TenureGroup'] = pd.cut(df['YearsAtCompany'], bins=[0, 2, 5, 10, 50],
                           labels=['0-2 years', '2-5 years', '5-10 years', '10+ years'])

# Converting Attrition to Binary 0 ot 1 for calulation
df['AttritionBinary'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# Calculate attrition impact score for each factor
# Impact = (High Risk Attrition Rate - Low Risk Attrition Rate) × Affected Population %

impact_analysis = []

# 1. OverTime
ot_yes_rate = df[df['OverTime']=='Yes']['AttritionBinary'].mean() * 100
ot_no_rate = df[df['OverTime']=='No']['AttritionBinary'].mean() * 100
ot_yes_pct = (df['OverTime']=='Yes').mean() * 100
impact_analysis.append({
    'Factor': 'OverTime',
    'High_Risk_Rate': ot_yes_rate,
    'Low_Risk_Rate': ot_no_rate,
    'Difference': ot_yes_rate - ot_no_rate,
    'Affected_Population_%': ot_yes_pct,
    'Impact_Score': (ot_yes_rate - ot_no_rate) * (ot_yes_pct / 100)
})

# 2. Job Satisfaction (Low vs High)
js_low_rate = df[df['JobSatisfaction']==1]['AttritionBinary'].mean() * 100
js_high_rate = df[df['JobSatisfaction']==4]['AttritionBinary'].mean() * 100
js_low_pct = (df['JobSatisfaction']==1).mean() * 100
impact_analysis.append({
    'Factor': 'JobSatisfaction (Low)',
    'High_Risk_Rate': js_low_rate,
    'Low_Risk_Rate': js_high_rate,
    'Difference': js_low_rate - js_high_rate,
    'Affected_Population_%': js_low_pct,
    'Impact_Score': (js_low_rate - js_high_rate) * (js_low_pct / 100)
})

# 3. Monthly Income (Bottom 25% vs Top 25%)
income_q1_rate = df[df['MonthlyIncome'] <= df['MonthlyIncome'].quantile(0.25)]['AttritionBinary'].mean() * 100
income_q4_rate = df[df['MonthlyIncome'] >= df['MonthlyIncome'].quantile(0.75)]['AttritionBinary'].mean() * 100
income_q1_pct = 25  # By definition
impact_analysis.append({
    'Factor': 'MonthlyIncome (Low)',
    'High_Risk_Rate': income_q1_rate,
    'Low_Risk_Rate': income_q4_rate,
    'Difference': income_q1_rate - income_q4_rate,
    'Affected_Population_%': income_q1_pct,
    'Impact_Score': (income_q1_rate - income_q4_rate) * 0.25
})


# 4. Years at Company (0-2 years vs 10+ years)
yac_low_rate = df[df['YearsAtCompany'] <= 2]['AttritionBinary'].mean() * 100
yac_high_rate = df[df['YearsAtCompany'] >= 10]['AttritionBinary'].mean() * 100
yac_low_pct = (df['YearsAtCompany'] <= 2).mean() * 100
impact_analysis.append({
    'Factor': 'YearsAtCompany (0-2)',
    'High_Risk_Rate': yac_low_rate,
    'Low_Risk_Rate': yac_high_rate,
    'Difference': yac_low_rate - yac_high_rate,
    'Affected_Population_%': yac_low_pct,
    'Impact_Score': (yac_low_rate - yac_high_rate) * (yac_low_pct / 100)
})

# 5. Work-Life Balance (Bad vs Best)
wlb_bad_rate = df[df['WorkLifeBalance']==1]['AttritionBinary'].mean() * 100
wlb_best_rate = df[df['WorkLifeBalance']==4]['AttritionBinary'].mean() * 100
wlb_bad_pct = (df['WorkLifeBalance']==1).mean() * 100
impact_analysis.append({
    'Factor': 'WorkLifeBalance (Bad)',
    'High_Risk_Rate': wlb_bad_rate,
    'Low_Risk_Rate': wlb_best_rate,
    'Difference': wlb_bad_rate - wlb_best_rate,
    'Affected_Population_%': wlb_bad_pct,
    'Impact_Score': (wlb_bad_rate - wlb_best_rate) * (wlb_bad_pct / 100)
})


# Create DataFrame and sort by Impact Score
impact_df = pd.DataFrame(impact_analysis).sort_values('Impact_Score', ascending=False)
print("\n=== TOP 5 ATTRITION DRIVERS BY IMPACT ===")
print(impact_df.to_string(index=False))

# Visualize
plt.figure(figsize=(12, 8))
plt.barh(impact_df['Factor'], impact_df['Impact_Score'], color='darkred')
plt.xlabel('Impact Score', fontsize=12)
plt.ylabel('Factor', fontsize=12)
plt.title('Top Factors Driving Attrition (Impact Score)', fontsize=16)
plt.tight_layout()
plt.savefig('attrition_impact_score.png')
plt.show()

# Save analysis
impact_df.to_csv('attrition_impact_analysis.csv', index=False)
```

```python
#Create Risk Profile

# Define high-risk criteria based on analysis
high_risk_criteria = (
    (df['OverTime'] == 'Yes') &
    (df['JobSatisfaction'] <= 2) &
    (df['YearsAtCompany'] <= 2) &
    (df['MonthlyIncome'] <= df['MonthlyIncome'].quantile(0.25))
)

# Calculate attrition rate for high-risk profile
high_risk_attrition = df[high_risk_criteria]['AttritionBinary'].mean() * 100
high_risk_count = high_risk_criteria.sum()
high_risk_pct = (high_risk_count / len(df)) * 100

print(f"\n=== HIGH-RISK EMPLOYEE PROFILE ===")
print(f"Criteria: OverTime=Yes, JobSatisfaction≤2, Tenure≤2 years, Income in bottom 25%")
print(f"Employees matching profile: {high_risk_count} ({high_risk_pct:.1f}% of workforce)")
print(f"Attrition rate for this profile: {high_risk_attrition:.1f}%")
print(f"vs Overall attrition rate: {df['AttritionBinary'].mean()*100:.1f}%")
print(f"Risk multiplier: {high_risk_attrition / (df['AttritionBinary'].mean()*100):.1f}x")
```

```python
#Detailed Profile Characteristics


# Compare high-risk vs overall workforce
print("\n=== PROFILE COMPARISON ===")

comparison_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction', 
                      'EnvironmentSatisfaction', 'WorkLifeBalance', 'DistanceFromHome']

comparison_data = []
for feature in comparison_features:
    high_risk_avg = df[high_risk_criteria][feature].mean()
    overall_avg = df[feature].mean()
    comparison_data.append({
        'Feature': feature,
        'High_Risk_Avg': high_risk_avg,
        'Overall_Avg': overall_avg,
        'Difference': high_risk_avg - overall_avg
    })

comparison_df = pd.DataFrame(comparison_data)
print(comparison_df.to_string(index=False))

# Visualize profile
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Age distribution
axes[0,0].hist([df[high_risk_criteria]['Age'], df['Age']], 
               label=['High Risk', 'Overall'], bins=20, alpha=0.7)
axes[0,0].set_title('Age Distribution')
axes[0,0].set_xlabel('Age')
axes[0,0].legend()

# Income distribution
axes[0,1].hist([df[high_risk_criteria]['MonthlyIncome'], df['MonthlyIncome']], 
               label=['High Risk', 'Overall'], bins=20, alpha=0.7, color=['red', 'blue'])
axes[0,1].set_title('Monthly Income Distribution')
axes[0,1].set_xlabel('Monthly Income')
axes[0,1].legend()

# Tenure distribution
axes[1,0].hist([df[high_risk_criteria]['YearsAtCompany'], df['YearsAtCompany']], 
               label=['High Risk', 'Overall'], bins=20, alpha=0.7, color=['red', 'green'])
axes[1,0].set_title('Years at Company Distribution')
axes[1,0].set_xlabel('Years at Company')
axes[1,0].legend()

# Job Satisfaction
js_high_risk = df[high_risk_criteria]['JobSatisfaction'].value_counts().sort_index()
js_overall = df['JobSatisfaction'].value_counts().sort_index()
x = np.arange(len(js_overall))
width = 0.35
axes[1,1].bar(x - width/2, js_high_risk.values, width, label='High Risk', color='red', alpha=0.7)
axes[1,1].bar(x + width/2, js_overall.values, width, label='Overall', color='blue', alpha=0.7)
axes[1,1].set_title('Job Satisfaction Distribution')
axes[1,1].set_xlabel('Job Satisfaction (1=Low, 4=High)')
axes[1,1].set_ylabel('Count')
axes[1,1].set_xticks(x)
axes[1,1].set_xticklabels(js_overall.index)
axes[1,1].legend()

plt.tight_layout()
plt.savefig('high_risk_profile_characteristics.png')
plt.show()
```
